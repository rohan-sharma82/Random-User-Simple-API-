```python
from randomuser import RandomUser
import pandas as pd

r=RandomUser()
```

```python
#using generate_users() function, we get a list of random 10 users.
user_list=r.generate_users(15)
user_list
```

```python
name=r.get_full_name()
for user in user_list:
  print(user.get_full_name(),"",user.get_email())
```

```python
#generating pictures of random users
for user in user_list:
  print(user.get_picture())


#getting passwords
print("Passwords\n")
for user in user_list:
  print(user.get_password())
a=[]
for user in user_list:
  a.append([user.get_postcode(),user.get_state(),user.get_city(),user.get_street()])
  df=pd.DataFrame(a,columns=["PostCode","State","City","stree"])
print(a)
df
```

```python
from collections import defaultdict

grouped_users = defaultdict(list)
for user in user_list:
    country = user.get_nat()  # 'get_nat' returns nationality code like 'US', 'GB'
    grouped_users[country].append(user.get_full_name())

# Print grouped names
for nat, names in grouped_users.items():
    print(f"{nat}: {names}")
```
